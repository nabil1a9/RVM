```python
import Levenshtein

words = ['FANTA','fanta', 'cocacola', 'délice', 'BOGA', 'apla', 'SCHWEPPES' , 'schweppes', 'boga', 'APLA', 'SPRITE','orangine', 'ORANGINA', 'FREEZ','freez','RED BULL', 'red bull','PEPSI', 'MR PEPPER' 'boga', 'pepsi', 'mr.pepper', 'sprite','SABRINE','SAFIA','FOURAT','MELLITI', 'delice','Tijen','Marwa', 'Janet', 'Dima']

def find_most_similar_words(word, words_list):
    similar_words = []
    min_distance = float('inf')
    for w in words_list:
              # Calculate the Levenshtein distance between the word and each word in the list
        distance = Levenshtein.distance(word, w)
        if distance < min_distance:
          # If the distance is smaller than the current minimum distance,
            # update the minimum distance and set the similar_words list to contain only the current word
            min_distance = distance
            similar_words = [w]
        elif distance == min_distance:
           # If the distance is equal to the current minimum distance,
            # append the current word to the similar_words list
            similar_words.append(w)
    return similar_words
```
